**Purpose**: This script will output a wiki page that can be used for listing books.

**Input**: Metadata csv file

**To Run**: Specify the start_noolaham_num number and execute the blocks 


```python
start_noolaham_num = 61001
end_noolaham_num = start_noolaham_num + 100
```

```python
# -*- coding: utf-8 -*-
# Load required libs
import os
import csv
```

```python
# Function used to calculate the bytes size of a unique string
def utf8len(s):
    return len(s.encode('utf-8'))
```

```python
# Main logic to create the wiki table
table_start = """
{{பட்டியல்கள் வார்ப்புரு}}
[[File:Under_Construction.jpg|center]]

{| class="wikitable" border="1" cellpadding="2" style="text-align:left" width="100%"
|-
! எண்
! நூல்/இதழ்
! ஆசிரியர்/காலம்
"""


batchfile = open('createdata.csv', 'rt')
try:
    reader = csv.reader(batchfile)
    next(reader, None)
    
    table = table_start
    for row in reader:
        noolaham_num = row[0].strip()
        noolaham_num_int = int(row[0])
        if not (noolaham_num_int >= start_noolaham_num and noolaham_num_int < end_noolaham_num):
            continue;

        title = row[2].strip()
        if(utf8len(title) > 255):
            print("Warning: Long Title " + noolaham_num + " - " + " - " + title + "\n") 

        date = row[3].strip()
        author = row[8].strip()
        table_entry ="|-\n"
        table_entry = table_entry + "|" +  noolaham_num + "\n"
        table_entry = table_entry + "|[[" +  title + "]]\n"
        
        if len(date) > 1 and len(author) > 1:
            table_entry = table_entry + "|" +  author + " / " + date + "\n"
        elif len(author) > 1:
            table_entry = table_entry + "|" +  author + "\n"
        elif len(date) > 1:
            table_entry = table_entry + "|" +  date + "\n"
        else:
            table_entry = table_entry + "| - \n"

        table = table + table_entry
finally:
     batchfile.close()
     
table_end = """|-

|-
|-
|}    
{{பட்டியல்கள்}}
"""
table = table + table_end
```

```python
# Output the result to a file
output_file_name = str(start_noolaham_num) + " _ " + str(end_noolaham_num) + "_listing.txt"
  
text_file = open(output_file_name, "w")
text_file.write(table)
text_file.close()
```
